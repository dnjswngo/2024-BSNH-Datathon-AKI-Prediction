```python
# 'KCD 코드' 값에서 NaN 또는 float 값 처리, 문자열로 변환 후 중복을 제거하고 쉼표로 연결
icd_codes = diag.groupby('등록번호')['KCD 코드'].apply(
    lambda x: ', '.join(sorted(set(str(code) for code in x if pd.notna(code))))
).reset_index()

# 'patients' 데이터프레임에 'KCD 코드' 열을 병합 
patients = patients.merge(icd_codes, on='등록번호', how='left')

# 결과 확인
patients.head()
```

```python
# diagnose = 1: 'N17' (AKI) 코드가 있을 경우
# diagnose = 2: 'N18' (CKD) 코드가 있을 경우
# diagnose = 3: 'N17'과 'N18' 두 코드가 모두 있을 경우
# diagnose = 0: 둘 다 해당하지 않는 경우

patients['diagnose'] = 0  # 기본값 0으로 설정

for index, row in patients.iterrows():
    has_aki = any(code == 'N17' for code in row['KCD 코드'])
    has_ckd = any(code == 'N18' for code in row['KCD 코드'])

    if has_aki and has_ckd:
        patients.at[index, 'diagnose'] = 3
    elif has_aki:
        patients.at[index, 'diagnose'] = 1
    elif has_ckd:
        patients.at[index, 'diagnose'] = 2
patients
```

```python
# set(bt['항목값'])
map = { # 이상한 값들 유형 나눠봄
 # ..이 2개 껴 있음
 '39..0' ,'38..0','38..1', '38..2','38..3', '37..0', '37..1', '37..2','37..3','37..4', '37..5', '37..6', '37..7',
 '37..8', '37..9', '38..4','36..', '36..0', '36..1', '36..2', '36..3', '36..4', '36..5', '36..6', '36..7', '36..8','36..9',
 '35..2','35..3', '35..4', '35..5', '35..6', '35..8',  '34..2', '34..7', '34..8',  '33..8',  '3..56', 
 
 # 뒤에 점 하나 더 붙음
 '36.3.',
 
 # 앞에 점 하나 더 붙음
 '.6.0',

# 이런 경우
 '35.735.5','35.335.3',

# 공백으로
 ' ',
 '  ',
 
# 마이너스 나 마이너스 값. 마이너스는 차피 BT에서 안나오니까 nan값 처리 대상으로 간주
 '-',
 '-0.4',
 
 # .으로만 구성
 '.',
 '...'
 }
import re
import numpy as np

# 이상한 값을 정리하는 함수 정의
def clean_value_v2(value):
# 값이 문자열인 경우에만 처리 (숫자형 값 처리 방지)
    if isinstance(value, str):
        # 공백 제거
        value = value.strip()

        # 문자열에 '.'만 있는 경우 또는 '-'가 포함된 경우 NaN으로 처리
        if all(char == '.' for char in value) or '-' in value:
            return np.nan

        # `..`을 `.`으로 바꾸기
        value = value.replace('..', '.')

        # '35.735.5' 같은 경우 첫 번째 숫자와 소수점만 남기고 이후 삭제
        value = re.sub(r'(\d+\.\d+).*', r'\1', value)

        # '.숫자' 같은 형식이면 숫자만 남기기
        if value.startswith('.'):
            value = value[1:]
    
    # 값이 숫자가 될 수 있는지 확인하고 변환, 그렇지 않으면 NaN 처리
    try:
        return float(value)
    except (ValueError, TypeError):
        return np.nan

# 'bt' 데이터프레임의 '항목값' 열에 함수 적용
bt['항목값'] = bt['항목값'].apply(clean_value_v2)
```

```python
# '등록번호'와 '간호기록 작성일자'별로 BT의 최소값, 최대값, 평균값 계산
bt_data = bt.groupby(['등록번호', '간호기록 작성일자']).agg(
    BT_min=('항목값', 'min'),
    BT_max=('항목값', 'max'),
    BT_mean=('항목값', 'mean')
).reset_index()

# 결과 확인
bt_data.head()
```

```python
pr_map = { '104104':'104','105103':'105','107107':'107','108108':'108', '120120':'120',
 
 # 이 친구들은 함수로 처리
#  '10-106', '110-120',
#  '110-121',
#  '110-140', '111-115','114-120', '120-160',
#  '120-175','128-140', '130-150','150-160','158-163','48-60','50-70','58-90','75-100',
#  '85-100', '87-100','88-150',
#  '88-90',
 
 '5963' :'59-63' ,
 ' ' : np.nan}
pr['항목값'] = pr['항목값'].map(pr_map).fillna(pr['항목값'])
```

```python
# 주어진 문자열 값에서 '-'가 포함된 경우 두 값의 평균으로 바꾸는 함수 정의
def convert_range_to_mean(value):
    if isinstance(value, str) and '-' in value:
        value = value.strip()
        try:
            # '-'로 나누어 두 숫자를 추출
            parts = value.split('-')
            # 두 숫자의 평균 계산
            mean_value = (float(parts[0]) + float(parts[1])) / 2
            return mean_value
        except (ValueError, IndexError):
            # 값이 숫자가 아니거나 '-'로 나누었을 때 문제가 생기면 None 반환
            return None
    # '-'가 없으면 원래 값 반환
        # 값이 숫자가 될 수 있는지 확인하고 변환, 그렇지 않으면 NaN 처리
    try:
        return float(value)
    except (ValueError, TypeError):
        return np.nan

# '-'가 포함된 값들을 함수로 처리
pr['항목값'] = pr['항목값'].apply(convert_range_to_mean)
```
